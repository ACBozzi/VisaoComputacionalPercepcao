Leia sobre Principal Component Analysis e veja se consegue algum resultado interessante neste dataset

```python
from sklearn import decomposition
from sklearn.preprocessing import StandardScaler
import mpl_toolkits.mplot3d 

# PCA com 3 componentes ( dimensões )

np.random.seed(5)

X = iris.drop('species', axis=1)
X_std = StandardScaler().fit_transform(X)

y = iris['species'].apply(lambda x: {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}[x])

fig = plt.figure(1, figsize=(4, 3))
plt.clf()

ax = fig.add_subplot(111, projection="3d", elev=48, azim=134)
ax.set_position([0, 0, 0.95, 1])


plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(X_std)
X = pca.transform(X_std)

for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        X[y == label, 0].mean(),
        X[y == label, 1].mean() + 1.5,
        X[y == label, 2].mean(),
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
    )

# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

plt.title("PCA com 3 componentes")
plt.show()
```

```python
# PCA com 2 componentes

X = iris.drop('species', axis=1)
X_std = StandardScaler().fit_transform(X)

y = iris['species']#.apply(lambda x: {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}[x])


sklearn_pca = decomposition.PCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(X_std)

with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
  
    for lab, col in zip(('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'),
                        ('blue', 'red', 'green')):
        plt.scatter(Y_sklearn[y==lab, 0],
                    Y_sklearn[y==lab, 1],
                    label=lab,
                    c=col)
  
    plt.title('PCA com 2 Componentes')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

Divida o dataset em 80% para treinamento e 20% para teste

Tente separar as classes com K-NN, analise resultados.

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
```

```python
treino, teste = train_test_split(iris, test_size=0.20, random_state=2021)

columns_treino = treino.select_dtypes(float).copy()
target_treino = treino['species'].copy()

columns_teste = teste.select_dtypes(float).copy()
target_teste = teste['species'].copy()
```

```python
knn = KNeighborsClassifier(n_neighbors=1)

#treinando o modelo
knn.fit(columns_treino, target_treino)
```

```python
pred = knn.predict(columns_teste)
print(f'Predição: {pred[0:5]}',end='\n\n')

acuracia = metrics.accuracy_score(pred, target_teste)
print(f'A acurácia é de {acuracia}')
```
