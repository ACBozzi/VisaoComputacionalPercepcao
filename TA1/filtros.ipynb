```python
#bibliotecas necessárias
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
#carregar as imagens
img1 = cv2.imread('/content/drive/My Drive/Visão e Percepção/imagens/family004.bmp')
```

```python
cv2_imshow(img1)
```

```python
# Filtro de média
blur = cv2.blur(img1, (5, 5)) # kernel de 5x5 pixels
cv2_imshow(blur)
```

```python
# Filtro gaussiano
gaussian = cv2.GaussianBlur(img1, (5, 5), 0) # kernel de 5x5 pixels
cv2_imshow(gaussian)
```

```python
# Filtro mediano
median = cv2.medianBlur(img1, 5) # kernel de 5x5 pixels
cv2_imshow(median)
```

```python
sobelx = cv2.Sobel(img1, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(img1, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.convertScaleAbs(sobelx) + cv2.convertScaleAbs(sobely)
cv2_imshow(sobel)
```

```python
laplacian = cv2.Laplacian(img1, cv2.CV_64F)
cv2_imshow(laplacian)
```

```python
blur = cv2.GaussianBlur(img1, (5, 5), 0)
unsharp = cv2.addWeighted(img1, 1.5, blur, -0.5, 0)
cv2_imshow(unsharp)
```

```python
canny = cv2.Canny(img1, 100, 200)
cv2_imshow(canny)
```
