Thresholding

Normal Thresholding

```python
#normal thresholding

# Carrega a imagem em escala de cinza
# img = cv2.imread('/content/drive/My Drive/imagens/bart010.bmp', 0)
img = cv2.imread('/content/drive/My Drive/Visão e Percepção/imagens/torre.webp',0)

# Aplica Normal Thresholding

#imagem binária em que os pixels são brancos se estiverem acima do limiar e pretos se estiverem abaixo.
limiar, img_bin = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

final_image = cv2.hconcat((img, img_bin))

# Exibe as imagens
cv2_imshow(final_image)
```

Automatic Thresholding

```python
#Automatic Thresholding


# Aplica Automatic Thresholding com o algoritmo OTSU
# esse algoritmo de limiarização automática determina o valor de limiar automaticamente com base na distribuição dos níveis de intensidade dos pixels na imagem.
limiar, img_bin = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

final_image = cv2.hconcat((img, img_bin))

# Exibe as imagens
cv2_imshow(final_image)
```

Multilevel thresholding

```python
#Multilevel thresholding

# Aplica Multilevel thresholding com 3 níveis de limiar
#os valores de limiar são calculados para 3 níveis e as imagens binárias são geradas com base nesses valores de limiar usando a função cv2.threshold() com cv2.THRESH_BINARY como tipo de thresholding.
limiares = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
limiar1 = limiares[0] / 3
limiar2 = limiares[0] * 2 / 3
img_bin = cv2.threshold(img, limiar1, 255, cv2.THRESH_BINARY)
img_bin1 = cv2.threshold(img, limiar2, 255, cv2.THRESH_BINARY)
img_bin2 = cv2.threshold(img, limiares[0], 255, cv2.THRESH_BINARY)

# Exibe as imagens
final_image = cv2.hconcat((img, img_bin[1]))
cv2_imshow(final_image)
```

```python
# Exibe as imagens
final_image = cv2.hconcat((img_bin1[1], img_bin2[1]))
cv2_imshow(final_image)
```

Local Thresholding

```python
#Local Thresholding

# Aplica Local Thresholding com o método de limiarização adaptativa Mean
img_bin = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)

# Exibe as imagens
final_image = cv2.hconcat((img, img_bin))

cv2_imshow(final_image)
```
